```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.rcdefaults()
fig, ax = plt.subplots()
census = pd.read_csv('Census.csv', encoding="ISO-8859-1")
census.head()
```

```python
laborforce = pd.read_csv('LaborForce.csv')
laborforce.head()
```

```python
organize = laborforce.groupby(['OCC_TITLE', 'TOT_EMP']).size()
organize
```

```python
plt.rcdefaults()
fig, ax = plt.subplots()

state = census["State"]
y_pos = np.arange(len(census["Population"]))

population = census["Population"]

ax.barh(y_pos, population, align='center', ecolor="black")

ax.set_yticks(y_pos)
ax.set_yticklabels(state)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Population')
ax.set_title('State Population')

plt.figure(figsize=(18, 18))
plt.show()
```
